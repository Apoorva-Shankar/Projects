```python
movie_title_lst = []
movie_score_lst = []

for movie in episode_titles:

    #Movie Title
    movie_title = movie.a.text
    print(f'movie_name: {movie_title}')
    

    try:
        #movie_score
            movie_score = movie.find(
            'div',
            attrs={'class': 'ipl-rating-widget'
                  }).find(
            'div',
            attrs={'class': 'ipl-rating-star small'
                  }).find(
            'span',
            attrs={'class': 'ipl-rating-star__rating'
                  }).text
    except:
        pass

    print(f'movie_score: {movie_score}')
    movie_title_lst.append(movie_title)
    movie_score_lst.append(movie_score)
    print('\n')

```

```python
df = pd.DataFrame({
    'title':movie_title_lst,
    'rating':movie_score_lst
})
df.to_csv('C:/Users/Apoorva Shankar/Dropbox/PC/Desktop/MQM Business Analytics/Fall Term 1/Data Science for Business/IMDB_Disney_Movie_Reviews_7.csv')
```
